```python
from google.colab import files
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/content/planilha_maiores_dividendos_2024.xlsx')
```

```python
# Faz uma cópia dos cabeçalhos atuais
new_headers = df.columns.to_list()
new_headers
```

```python
for i, header in enumerate(new_headers):
    # Verifica se o valor na linha 0 não é NaN
    if pd.notnull(df.iloc[0, i]):
        # Anexa o valor da linha 0 ao cabeçalho da coluna
        new_headers[i] = f"{header} {df.iloc[0, i]}".strip()
```

```python
# Atualiza os cabeçalhos da DataFrame
df.columns = new_headers
df
```

```python
# Remove a primeira linha de dados
df = df.drop(index=0)
df
```

```python
# Reseta o índice da DataFrame
df.reset_index(drop=True, inplace=True)

# Exibe os primeiros elementos para conferência
print(df.head())
```

```python

# Function to remove currency and percentage signs and convert to float
def remove_currency_percent(val):
    if isinstance(val, str):
        # Remove currency and percentage signs
        val = val.replace('R$', '').replace('%', '').replace('Bilhões', '').replace(',', '.').strip()
        # Convert to float
        try:
            return float(val)
        except ValueError:
            return val  # Return the value as is if it cannot be converted to float
    return val
```

```python
# Apply the function to each column except 'Unnamed: 0'
for col in df.columns[1:]:
    df[col] = df[col].apply(remove_currency_percent)
```

```python

# Now let's convert 'VALOR DE MERCADO' assuming 'Bilhões' translates to billion
df['VALOR DE MERCADO'] = df['VALOR DE MERCADO'].apply(lambda x: x if isinstance(x, str) else x * 1e9)
```

```python
# Tratar '-' como NaN
df.replace('-', np.nan, inplace=True)
```

```python
# Show the transformed DataFrame
print(df)
```

```python
# Salva o DataFrame como um arquivo Excel
df.to_excel('planilha_maiores_dividendos_2024_atualizada.xlsx', index=False)

# Informa ao usuário que o arquivo foi salvo com sucesso
print('DataFrame salvo como um arquivo Excel com sucesso!')
```
